```python
def test_calculate_norm():
    # 1. הגדרת פרמטרים לבדיקה
    w0 = 1.0
    tau = 1.0
    theta_p = -0.5  # רדיאנים
    
    # 2. יצירת אובייקט הפולס (נניח שזה השם של המחלקה שלך)
    pulse = UltrafastPulse(w0=w0, tau=tau, zeta=0)    
    # אובייקט הטרנספורמציה שמכיל את הזווית
    class Transform:
        def __init__(self, angle):
            self.theta_p = angle
    
    transform = Transform(theta_p)
    
    # 3. הרצת הפונקציה החדשה
    calculated_norm = pulse.calculate_norm(transform)
    
    # 4. חישוב הערך הצפוי אנליטית (לפי הנוסחה למעלה)
    import numpy as np
    expected_norm = (np.pi**1.5 * w0**2 * tau * np.abs(np.cos(theta_p))) / (2 * np.sqrt(2))
    
    print(f"Calculated: {calculated_norm}")
    print(f"Expected:   {expected_norm}")
    
    # 5. בדיקת דיוק (Assert)
    assert np.isclose(calculated_norm, expected_norm, rtol=1e-5)
    print("✅ Test Passed!")

# הרצה
test_calculate_norm()
```

```python
# הגדרת הקשר הגיאומטרי
z_surf = -pulse.sp_x * sp.tan(pulse.sp_theta_p)
xp_sym = pulse.sp_x * sp.cos(pulse.sp_theta_p) - z_surf * sp.sin(pulse.sp_theta_p)

# קבלת המעטפת והעוצמה
envelope_sym = pulse.get_envelope(xp_sym, pulse.sp_y, pulse.sp_t, library=sp)
# במקום Abs, נשתמש בכפל בצמוד ופישוט ממוקד
intensity_sym = (envelope_sym * sp.conjugate(envelope_sym)).expand()
intensity_sym = sp.simplify(intensity_sym)
print("Intensity expression is ready!")
sp.pprint(intensity_sym)
```

```python
# אינטגרציה סימבולית - זה החלק שעשוי לקחת זמן
norm_analytical = sp.integrate(intensity_sym, 
                               (pulse.sp_x, -sp.oo, sp.oo), 
                               (pulse.sp_y, -sp.oo, sp.oo), 
                               (pulse.sp_t, -sp.oo, sp.oo))

print("\nAnalytical Norm Result:")
sp.pprint(norm_analytical)
```

```python
import sympy as sp

# הגדרת הביטוי המקורי
intensity_expr = sp.exp(-2 * pulse.sp_t**2 / pulse.sp_tau**2) * \
                 sp.exp(-2 * pulse.sp_y**2 / pulse.sp_w0**2) * \
                 sp.exp(-2 * pulse.sp_x**2 / (pulse.sp_w0**2 * sp.cos(pulse.sp_theta_p)**2))

print("--- אינטגרציה מדורגת עם חילוץ ענפים (Robust Step-by-Step) ---")

try:
    # שלב 1: אינטגרל לפי x
    print("שלב 1: מחשב אינטגרל לפי x...")
    res_x_raw = sp.integrate(intensity_expr, (pulse.sp_x, -sp.oo, sp.oo))
    
    # חילוץ הענף הרלוונטי אם מדובר ב-Piecewise
    res_x = res_x_raw.args[0][0] if isinstance(res_x_raw, sp.Piecewise) else res_x_raw
    print("שלב 1 הושלם.")
    
    # שלב 2: אינטגרל לפי y
    print("שלב 2: מחשב אינטגרל לפי y...")
    res_y_raw = sp.integrate(res_x, (pulse.sp_y, -sp.oo, sp.oo))
    
    # שוב חילוץ למקרה שנוצר Piecewise חדש
    res_y = res_y_raw.args[0][0] if isinstance(res_y_raw, sp.Piecewise) else res_y_raw
    print("שלב 2 הושלם.")
    
    # שלב 3: אינטגרל לפי t
    print("שלב 3: מחשב אינטגרל לפי t...")
    final_norm_raw = sp.integrate(res_y, (pulse.sp_t, -sp.oo, sp.oo))
    
    # חילוץ סופי
    final_norm = final_norm_raw.args[0][0] if isinstance(final_norm_raw, sp.Piecewise) else final_norm_raw
    
    print("\n--- התוצאה הסופית של הנורמה: ---")
    sp.pprint(final_norm)

except Exception as e:
    print(f"\nהתהליך נכשל: {e}")
```
