# CNN SNR results — plot only

Load precomputed CNN SNR sweeps (`.npz`) and PCGPA cache, then plot comparison curves.
No training here — run `scripts/train_trace_to_pulse.py` and the eval cell in `reconstruction_snr_experiments.ipynb` first.

```python
from pathlib import Path

import numpy as np

from evaluate_cnn import load_cnn_sweep, plot_standard_cnn_vs_pcgpa_suite

SNR_SWEEP_DB = np.arange(-10.0, 31.0, 5.0)
PCGPA_SWEEP_CACHE = "checkpoints/pcgpa_snr_sweep.npz"
PCGPA_LABEL = "PCGPA (best amb.)"

CNN_SWEEP_FILES = [
    "checkpoints/baseline_2k_cnn_sweep.npz",
    # "checkpoints/large_60k_cnn_sweep.npz",
]

cnn_results = [load_cnn_sweep(p) for p in CNN_SWEEP_FILES if Path(p).is_file()]
if not cnn_results:
    raise FileNotFoundError("No CNN sweep .npz found — run eval in reconstruction_snr_experiments.ipynb first.")

z = np.load(PCGPA_SWEEP_CACHE, allow_pickle=False)
assert np.array_equal(z["snr_sweep_db"], SNR_SWEEP_DB), "PCGPA cache SNR grid mismatch"

plot_standard_cnn_vs_pcgpa_suite(
    SNR_SWEEP_DB,
    cnn_results,
    pcgpa_sim_m=z["pcgpa_sim_m"],
    pcgpa_sim_s=z["pcgpa_sim_s"],
    pcgpa_l1_m=z["pcgpa_l1_m"],
    pcgpa_l1_s=z["pcgpa_l1_s"],
    pcgpa_label=PCGPA_LABEL,
)
```
